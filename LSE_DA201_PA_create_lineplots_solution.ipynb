Canopy is a new boutique streaming company that is looking to create an app that provides recommendations based on the last movie you watched. This analysis uses the `movies.csv` and `ott.xlsx` data sets. Based on the available information, in this activity you will:

- understand what the data says by visualising it with lineplots.

## 1. Import the libraries

```python
# Import necessary libraries.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 2. Import Excel file

```python
# Load the Excel data using pd.read_excel.
ott = pd.read_excel('ott.xlsx')

# View the columns.
print(ott.columns)
```

## 3. Import CSV file

```python
# Load the CSV data using pd.read_csv.
movies = pd.read_csv('movies.csv')

print(movies.columns)
```

## 4. Validate the DataFrames

```python
# Data imported correctly?
print(movies.head())
print(movies.shape)
print(movies.dtypes)
```

```python
# Data imported correctly.
print(ott.head())
print(ott.dtypes)
print(ott.shape)
```

## 5. Combine the two DataFrames

```python
# Merge the two DataFrames.
mov_ott = pd.merge(movies, ott, how='left', on = 'ID')

# View the DataFrame.
print(mov_ott.shape)
mov_ott.head()
```

## 6. Create a lineplot

```python
# Create a simple lineplot.
sns.lineplot(x='Year', y='IMDb', data=mov_ott)
```

```python
# Create a simple lineplot.
sns.lineplot(x='Year', y='IMDb', data=mov_ott, ci=None)
```

```python
# Create lineplots with specification.
sns.lineplot(x = 'Year', y = 'IMDb',
             data=mov_ott[mov_ott['Age'].isin(['16+', '18+'])],
             hue ='Age')
```

```python
# Create lineplots with specification.
sns.lineplot(x = 'Year', y = 'IMDb',
             data=mov_ott[mov_ott['Age'].isin(['16+', '18+'])],
             hue ='Age', ci=None)
```
